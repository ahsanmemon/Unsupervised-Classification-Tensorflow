```python
from PIL import Image, ImageEnhance, ImageOps
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import ndimage

from scan_tf.utils import augmentations
```

```python
# Load an image
im = Image.open(r"D:\Personal\AutoAugment-master\AutoAugment-master\images/im_0001.png")
plt.imshow(im)
```

```python
f, axarr = plt.subplots(4, 4, figsize=(15, 15), gridspec_kw={"wspace": 0, "hspace": 0})
for idx, ax in enumerate(f.axes):
    augmentation_name = np.random.choice(list(augmentations.func_pil.keys()))
    p = np.random.uniform(augmentations.strong_augment_bounds[augmentation_name][0], augmentations.strong_augment_bounds[augmentation_name][1])
    title = f"{augmentation_name} ({p:.2f})"
    augmented_im = augmentations.func_pil[augmentation_name](im, p)
    ax.imshow(augmented_im)
    ax.axis("off")
    ax.set_title(title)
plt.show()
```

```python
# Strong data augmentation
f, axarr = plt.subplots(4, 4, figsize=(15, 15), gridspec_kw={"wspace": 0, "hspace": 0})
for idx, ax in enumerate(f.axes):
    augmented_im = augmentations.strong_augmentation(im)
    ax.imshow(augmented_im)
    ax.axis("off")
plt.show()
```
